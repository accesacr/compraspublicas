<ul>

    <li>284 providers (~0.9%) appear with zero category ids </li>
    <li>Provider 0301870393 appears with category 10161501 in profile but category is not present in the search page of each goods and services</li>

    <li> All providers that have categories in sicop_provider_categorylevel4_map also have its category name in sicop_categorylevel4 </li>
</ul>

```sql
SELECT COUNT(*) FROM [cleandata].[sicop_provider] AS pro
```

```sql
SELECT 
    pro.Cedula
    ,cmap.category_id
FROM [cleandata].[sicop_provider] AS pro
LEFT OUTER JOIN [cleandata].[sicop_provider_categorylevel4_map] AS cmap ON pro.Cedula = cmap.provider_id
WHERE cmap.provider_id IS NULL
OR COALESCE(cmap.category_id, '') = ''
```

```sql
SELECT * FROM cleandata.sicop_categorylevel4 WHERE category_id LIKE '%10161501%'
```

```sql
SELECT 
    pro.Cedula
    ,cmap.category_id
    ,cat.category_name
FROM [cleandata].[sicop_provider] AS pro
LEFT OUTER JOIN [cleandata].[sicop_provider_categorylevel4_map] AS cmap ON pro.Cedula = cmap.provider_id
LEFT OUTER JOIN cleandata.sicop_categorylevel4 AS cat ON cmap.category_id = cat.category_id
WHERE cmap.provider_id IS NOT NULL
AND COALESCE(cmap.category_id, '') <> ''
AND (cat.category_id IS NULL
OR COALESCE(cat.category_name, '') = '')
```

<h1>Categories in mapping table with problems</h1>

<ul>
    <li>In total we have 738'355 distinct category mappings regardless of any problem</li>
    <li>All category_id in mapping table can find it's category name in category table</li>
    <li>29'513 mappings in category map table (~4%) can not find its provider in provider table, happens for 1938 providers (~6%)</li>
    
</ul>

```sql
SELECT count(*) FROM [cleandata].[sicop_provider_categorylevel4_map] AS cmap
```

```sql
SELECT 
    cmap.category_id
FROM [cleandata].[sicop_provider_categorylevel4_map] AS cmap 
WHERE COALESCE(cmap.provider_id, '') = ''
```

```sql
SELECT 
    cmap.category_id
    ,cat.category_name
FROM [cleandata].[sicop_provider_categorylevel4_map] AS cmap
LEFT OUTER JOIN cleandata.sicop_categorylevel4 AS cat ON cmap.category_id = cat.category_id
WHERE cmap.provider_id IS NOT NULL
AND COALESCE(cmap.category_id, '') <> ''
AND (cat.category_id IS NULL
OR COALESCE(cat.category_name, '') = '')
```

```sql
SELECT 
    cmap.provider_id
FROM [cleandata].[sicop_provider_categorylevel4_map] AS cmap 
LEFT OUTER JOIN [cleandata].[sicop_provider] AS pro ON pro.Cedula = cmap.provider_id
WHERE pro.Cedula IS NULL
```

```sql
SELECT 
    COUNT(DISTINCT cmap.provider_id) AS provider_id_count
FROM [cleandata].[sicop_provider_categorylevel4_map] AS cmap 
LEFT OUTER JOIN [cleandata].[sicop_provider] AS pro ON pro.Cedula = cmap.provider_id
WHERE pro.Cedula IS NULL
```
