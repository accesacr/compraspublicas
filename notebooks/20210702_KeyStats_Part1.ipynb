<h1> Participaciones por Procedimiento </h1>

```sql
SELECT 
    COUNT(DISTINCT procedure_num) AS Total_Procedimientos
    ,COUNT(*) AS Total_Participaciones
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) AS Promedio_Participaciones_x_Procedimiento
FROM cleandata.sicop_procedure_participations_per_provider

SELECT 
    COUNT(DISTINCT procedure_num) AS Total_Procedimientos
    ,COUNT(*) AS Total_Participaciones
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) AS Promedio_Participaciones_x_Procedimiento
    ,COUNT(DISTINCT Institución) AS Total_Instituciones
    ,CAST(COUNT(DISTINCT procedure_num) AS decimal) / CAST(COUNT(DISTINCT Institución) AS decimal) AS Promedio_Procedimientos_x_Institucion
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT Institución) AS decimal) AS Promedio_Participaciones_x_Institucion
    
FROM cleandata.sicop_procedure_participations_per_provider as pprov
INNER JOIN cleandata.sicop_procedure AS pro ON pprov.procedure_num = pro.Número_procedimiento

SELECT 
    YEAR(max_opening_act_date) AS Year_Max_Opening_Act_Date
    ,COUNT(DISTINCT procedure_num) AS Total_Procedimientos
    ,COUNT(*) AS Total_Participaciones
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) AS Promedio_Participaciones_x_Procedimiento
FROM cleandata.sicop_procedure_participations_per_provider
WHERE YEAR(max_opening_act_date) < 2021
GROUP BY YEAR(max_opening_act_date)
ORDER BY YEAR(max_opening_act_date) DESC



-- ;WITH cte_procedure AS (
--     SELECT
--         pc.Max_Fecha_Notificacion
--         ,p.*
--     FROM cleandata.sicop_procedure AS p
--     INNER JOIN (
--         SELECT 
--             pci.Número_procedimiento
--             ,MAX(pci.Fecha_notificación) AS Max_Fecha_Notificacion
--         FROM cleandata.sicop_procedure_contract AS pci
--         GROUP BY pci.Número_procedimiento
--     ) AS pc ON p.Número_procedimiento = pc.Número_procedimiento
-- )
SELECT 
    --YEAR(Max_Fecha_Notificacion) AS Max_Fecha_Notificacion_Contrato_Procedimiento
    YEAR(max_opening_act_date) AS Year_Max_Opening_Act_Date
    ,COUNT(DISTINCT procedure_num) AS Total_Procedimientos
    ,COUNT(*) AS Total_Participaciones
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) AS Promedio_Participaciones_x_Procedimiento
    ,COUNT(DISTINCT Institución) AS Total_Instituciones
    ,CAST(COUNT(DISTINCT procedure_num) AS decimal) / CAST(COUNT(DISTINCT Institución) AS decimal) AS Promedio_Procedimientos_x_Institucion
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT Institución) AS decimal) AS Promedio_Participaciones_x_Institucion
    
FROM cleandata.sicop_procedure_participations_per_provider as pprov
INNER JOIN cleandata.sicop_procedure  AS pro ON pprov.procedure_num = pro.Número_procedimiento
GROUP BY YEAR(max_opening_act_date)
ORDER BY YEAR(max_opening_act_date) DESC
--GROUP BY YEAR(Max_Fecha_Notificacion)
--ORDER BY YEAR(Max_Fecha_Notificacion) DESC
```

```sql
SELECT 
    pro.Institución
    ,COUNT(DISTINCT procedure_num) AS Total_Procedimientos
    ,COUNT(*) AS Total_Participaciones
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) AS Promedio_Participaciones_x_Procedimiento
FROM cleandata.sicop_procedure_participations_per_provider as pprov
INNER JOIN cleandata.sicop_procedure AS pro ON pprov.procedure_num = pro.Número_procedimiento
GROUP BY pro.Institución
ORDER BY CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) DESC
```

```sql
SELECT 
    Institución
    ,ROW_NUMBER() OVER (ORDER BY COUNT(DISTINCT procedure_num) DESC) AS Rank_Institucion_Total_Procedimientos
    ,ROW_NUMBER() OVER (ORDER BY COUNT(*) DESC) AS Rank_Institucion_Total_Participaciones
    ,ROW_NUMBER() OVER (ORDER BY CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) DESC) AS Rank_Institucion_Promedio_Participaciones_x_Procedimiento
    ,ROW_NUMBER() OVER (ORDER BY CAST(COUNT(DISTINCT procedure_num) AS decimal) / CAST(COUNT(DISTINCT Institución) AS decimal) DESC) AS Rank_Institucion_Promedio_Procedimientos_x_Institucion
    ,COUNT(DISTINCT procedure_num) AS Total_Procedimientos
    ,COUNT(*) AS Total_Participaciones
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) AS Promedio_Participaciones_x_Procedimiento
FROM cleandata.sicop_procedure_participations_per_provider as pprov
INNER JOIN cleandata.sicop_procedure AS pro ON pprov.procedure_num = pro.Número_procedimiento
GROUP BY Institución
ORDER BY ROW_NUMBER() OVER (ORDER BY CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) DESC) ASC

SELECT 
    YEAR(max_opening_act_date) AS Year_Max_Opening_Act_Date
    ,Institución
    ,ROW_NUMBER() OVER (ORDER BY COUNT(DISTINCT procedure_num) DESC) AS Rank_Institucion_Total_Procedimientos
    ,ROW_NUMBER() OVER (ORDER BY COUNT(*) DESC) AS Rank_Institucion_Total_Participaciones
    ,ROW_NUMBER() OVER (ORDER BY CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) DESC) AS Rank_Institucion_Promedio_Participaciones_x_Procedimiento
    ,ROW_NUMBER() OVER (ORDER BY CAST(COUNT(DISTINCT procedure_num) AS decimal) / CAST(COUNT(DISTINCT Institución) AS decimal) DESC) AS Rank_Institucion_Promedio_Procedimientos_x_Institucion
    ,YEAR(max_opening_act_date) AS Year_Max_Opening_Act_Date
    ,COUNT(DISTINCT procedure_num) AS Total_Procedimientos
    ,COUNT(*) AS Total_Participaciones
    ,CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) AS Promedio_Participaciones_x_Procedimiento
FROM cleandata.sicop_procedure_participations_per_provider as pprov
INNER JOIN cleandata.sicop_procedure  AS pro ON pprov.procedure_num = pro.Número_procedimiento
GROUP BY YEAR(max_opening_act_date), Institución
ORDER BY YEAR(max_opening_act_date) DESC, ROW_NUMBER() OVER (ORDER BY CAST(COUNT(*) AS decimal) / CAST(COUNT(DISTINCT procedure_num) AS decimal) DESC) ASC


```
